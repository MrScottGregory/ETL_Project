```python
# import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

## Vehicle Registration Data

```python
# read in vehicle registration CSV
reg_csv_file = "Vehicle__Snowmobile__and_Boat_Registrations.csv"
vehicle_reg_df = pd.read_csv(reg_csv_file)
```

```python
# explore the dataframe: review list of columns
vehicle_reg_df.columns
```

```python
# explore the dataframe: sample the top of the df
vehicle_reg_df.head()
```

```python
# explore the dataframe: determine how many Record Types exist
vehicle_reg_df.groupby(['Record Type']).count()
```

```python
# retain only those rows with a record type VEH (we only want vehicles, vs boats etc)
# note that record type has 4 characters, so we need to leave a space after VEH
veh_reg_df = vehicle_reg_df[vehicle_reg_df["Record Type"]== "VEH "]
veh_reg_df.tail()
```

```python
# drop columns not relevant to our analysis
registration_df = veh_reg_df.drop(columns=['Record Type',
                                           'Fuel Type', 
                                           'Maximum Gross Weight', 
                                           'Passengers',
                                           'Scofflaw Indicator'])
```

```python
# we want to clean up "Body Type" for joining with other data
# let's start by reviewing how many body types exist
# decode the body types here: https://data.ny.gov/api/views/w4pv-hbkt/files/AUsdC2Y0iEymGyebFASIjDxZ7irrm1-_yS-o9qFzWTQ?download=true&filename=NYSDMV_VehicleSnowmobileAndBoat_Registration_Data%20Dictionary.pdf
registration_df["Body Type"].value_counts()
```

```python
# we notice that there are still some boats and snowplows, so let's eliminate those
registration_df = registration_df[(registration_df['Body Type'] != 'TRAV') & 
                                 (registration_df['Body Type'] != 'BOAT')]
```

```python
# let's aggregate some of the body types together and convert to readable names
# remember, values have 4 characters, so:
# we need to add a space after any 3 character value, and let's rename so none are left
replace_values = {'4DSD' : 'SEDAN', '2DSD' : 'SEDAN', 'SEDN' : 'SEDAN',
                  'PICK' : 'PASS_TRUCK', 'VAN ' : 'PASS_TRUCK', 'LTRL' : 'PASS_TRUCK', 'SWT ' : 'PASS_TRUCK',
                  '????' : 'OTHER', 'RPLC' :'OTHER', 'CUST' : 'OTHER',
                  'AMBU' : 'EMERGENCY', 'FIRE' : 'EMERGENCY', 'DCOM' : 'EMERGENCY',
                  'CMIX' : 'MACHINE', 'P/SH' : 'MACHINE', 'S/SP' : 'MACHINE', 'SN/P' : 'MACHINE', 'EMVR' : 'MACHINE', 'W/DR' : 'MACHINE', 'W/SR': 'MACHINE', 'MCC ': 'MACHINE', 'FPM ' : 'MACHINE', 'LOCO' : 'MACHINE', 'RBM ' : 'MACHINE', 'RD/S' : 'MACHINE',
                  'DELV' : 'COMM_TRUCK', 'DUMP' : 'COMM_TRUCK', 'UTIL' : 'COMM_TRUCK', 'TRAC' : 'COMM_TRUCK', 'FLAT' : 'COMM_TRUCK', 'TRLR' : 'COMM_TRUCK', 'STAK' : 'COMM_TRUCK', 'TANK' : 'COMM_TRUCK', 'TOW ' : 'COMM_TRUCK', 'REFG' : 'COMM_TRUCK', 'TR/C' : 'COMM_TRUCK', 'SEMI' : 'COMM_TRUCK', 'T/CR' : 'COMM_TRUCK', 'H/TR' : 'COMM_TRUCK', 'R/RD' : 'COMM_TRUCK', 'TR/E' : 'COMM_TRUCK', 'POLE' : 'COMM_TRUCK',
                  'HRSE' : 'HEARSE', 'H/IN' : 'HEARSE',
                  'H/WH' : 'HOUSE_ON_WHEELS',
                  'LIM ' : 'TAXI',
                  'LSV ' : 'LOW_SPEED_VEH', 'LSVT': 'LOW_SPEED_VEH',
                  'MCY ' : 'CYCLES', 'MOPD' : 'CYCLES',
                  'SUBN' : 'SUBURBAN',
                  'ATV ' : 'ATV',
                  'BUS ' : 'BUS'
                 }
new_registration_df = registration_df.replace({'Body Type': replace_values}) 
```

```python
# review updated list of body types
new_registration_df['Body Type'].value_counts()
```

```python
new_registration_df.head()
```

```python
# review column names  
new_registration_df.columns
```

```python
# since the collision data is just for NYC, we want to narrow this data to NYC as well
# currently we have data for all counties in NY state
new_registration_df['County'].value_counts()
```

```python
# through trail and error we learned there are trailing blank characters in the county values
# so let's remove any trailing space with .str.strip() 
new_registration_df['County'] = new_registration_df['County'].str.strip()
```

```python
# now let's retain only those rows with data from NYC boroughs
nyc_registration_df = new_registration_df[(new_registration_df["County"]== "QUEENS") |
                                          (new_registration_df["County"]== "KINGS") |
                                          (new_registration_df["County"]== "NEW YORK") |
                                          (new_registration_df["County"]== "BRONX") |
                                          (new_registration_df["County"]== "RICHMOND")]

nyc_registration_df['County'].value_counts()
```

```python
# finally, let's rename our columns in snake-case
nyc_registration_df.columns = ['id', 'registration_class', 'city', 'state', 'zip', 'county', 
                               'model_year','make', 'body_type', 'unladen_weight', 
                               'reg_valid_date','reg_expiration_date', 'color', 
                               'suspension_indicator','revocation_indicator']
```

```python
nyc_registration_df.head()
```

```python
# create database connection
connection_string = "postgres:chumwater@localhost:5432/ny_vehicles_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm table exists
engine.table_names()
```

```python
# load dataframe into database
nyc_registration_df.to_sql(name='nyc_veh_registration', con=engine, if_exists='append', index=False)
```

```python
# output dataframe to CSV
nyc_registration_df.to_csv("nyc_registration.csv", index=True, header=True)
```

## Vehicle Collision Data

```python
# read in vehicle coliision CSV
collision_csv_file = "NYPD_Motor_Vehicle_Collisions.csv"
vehicle_accident_df = pd.read_csv(collision_csv_file)
```

```python
vehicle_accident_df.head()
```

```python
# analyze the data
vehicle_accident_df.info()
```

```python
# rename the columns
vehicle_accident_df.columns = ['date', 'time', 'borough','zip_code','lat', 'long', 'location', 'on_street_name',
                           'cross_street_name','off_street_name','num_persons_injured','num_persons_killed',
                           'num_pedestrians_injured','num_pedestrians_killed','num_cyclist_injured','num_cyclist_killed',
                           'num_motorist_injured','num_motorist_killed','contrib_fact_one','contrib_fact_two',
                           'contrib_fact_three','contrib_fact_four','contrib_fact_five','unique_key',
                           'veh_type_1', 'veh_type_2','veh_type_3','veh_type_4', 'veh_type_5']

vehicle_accident_df.head()
```

```python
# retain only the columns we need for the analysis
veh_accident_df = vehicle_accident_df[['date', 'time', 'borough','zip_code', 'lat','long','location','num_persons_injured',
                                       'num_persons_killed','num_pedestrians_injured','num_pedestrians_killed',
                                       'num_cyclist_injured','num_cyclist_killed','num_motorist_injured',
                                       'num_motorist_killed','contrib_fact_one', 'unique_key','veh_type_1']]

veh_accident_df.head()
```

```python
# rearrange columns so unique key is first
veh_accident_df = vehicle_accident_df[['unique_key','date', 'time', 'borough','zip_code', 'lat','long','location','num_persons_injured',
                                       'num_persons_killed','num_pedestrians_injured','num_pedestrians_killed',
                                       'num_cyclist_injured','num_cyclist_killed','num_motorist_injured',
                                       'num_motorist_killed','contrib_fact_one','veh_type_1']]
veh_accident_df.head()
```

```python
# rename unique key as id
veh_accident_df = veh_accident_df.rename(columns={"unique_key": "id"})
veh_accident_df.head()
```

```python
# remove rows with blanks
veh_accident_df = veh_accident_df.dropna()
```

```python
# to clean up vehicle type, let's start by retaining only the top 30 types
veh_accident_df['veh_type_1'].value_counts()
```

```python
# filter out the types that didn't make the top 30
veh_accident_df = veh_accident_df.groupby('veh_type_1').filter(lambda x: len(x) > 588)
veh_accident_df.info()
```

```python
# review current vehicle types
veh_accident_df['veh_type_1'].value_counts()
```

```python
# now let's aggregate and rename these types to be consistent with the registration table
replace_values = {'SPORT UTILITY / STATION WAGON':'SUBURBAN', 
                  'Station Wagon/Sport Utility Vehicle':'SUBURBAN',
                  'Taxi':'TAXI','VAN':'PASS_TRUCK',
                  'PICK-UP TRUCK': 'PASS_TRUCK', 
                  'Bus':'BUS','SMALL COM VEH(4 TIRES) ':'COMM_TRUCK',
                  'LARGE COM VEH(6 OR MORE TIRES)':'COMM_TRUCK',
                  'LIVERY VEHICLE':'TAXI',
                  'Pick-up Truck':'PASS_TRUCK',
                  'Box Truck':'COMM_TRUCK', 
                  'MOTORCYCLE':'CYCLES',
                  'BICYCLE':'CYCLES', 
                  'AMBULANCE':'EMERGENCY',
                  'TK':'COMM_TRUCK', 
                  'BU':'BUS', 
                  'Van':'PASS_TRUCK', 
                  'Bike':'CYCLES','Tractor Truck Diesel':'COMM_TRUCK',
                  'FIRE TRUCK':'EMERGENCY', 
                  'Motorcycle':'CYCLES', 
                  'Dump':'COMM_TRUCK', 
                  'Convertible':'CONV',
                  'Sedan':'SEDAN',  
                  'UNKNOWN':'OTHER'    
                 }

new_accident_df = veh_accident_df.replace({'veh_type_1': replace_values})
```

```python
new_accident_df.head()
```

```python
# create database connection
connection_string = "postgres:chumwater@localhost:5432/ny_vehicles_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm table exists
engine.table_names()
```

```python
# load dataframe into database
new_accident_df.to_sql(name='nyc_veh_collisions', con=engine, if_exists='append', index=False)
```

```python
# output dataframe to CSV
new_accident_df.to_csv("nyc_veh_collisions.csv", index=True, header=True)
```
